## The Chaotic Duffing Oscillator

The system includes dissipation and a periodic external forcing. The differential equations:

\begin{align*}
\dot{u} &= v, \\
\dot{v} &= -\gamma v + u - u^3 + \varepsilon \sin t,
\end{align*}

where $u$ and $v$ represent the coordinate and velocity, respectively, $\gamma$ is the damping coefficient, $\varepsilon$ is the amplitude of the external periodic force.

```python
from duffing import ChaoticDuffingSystem
from models import BaselineNN, HamiltonianNN
from visualizer import setup_plot_style, plot_poincare_sections
import numpy as np
import torch

setup_plot_style()
```

```python
params_values = ((0, 0), (0, 0.01), (0, 0.05), (0, 0.1), (0.15, 0.3))  # (gamma, epsilon)
simulations = []

for gamma, epsilon in params_values:
    print(f"Training models for gamma = {gamma}, epsilon = {epsilon}")

    system = ChaoticDuffingSystem(gamma=gamma, epsilon=epsilon)
    train_loader, test_loader = system.generate_dataset()

    bnn = BaselineNN(input_dim=4)
    optimizer = torch.optim.Adam(bnn.parameters(), lr=1e-3)
    bnn.train_model(train_loader, test_loader, optimizer)

    hnn = HamiltonianNN(input_dim=4)
    optimizer = torch.optim.Adam(hnn.parameters(), lr=1e-3)
    hnn.train_model(train_loader, test_loader, optimizer)

    simulations.append((system, bnn, hnn))
```

```python
u0_values = (
    [-np.sqrt(2) - 0.01, -np.sqrt(2)] +
    [-1.4, -1.35, -1.3, -1.2, -1.1, -1.0] +
    [0] +
    [1.0, 1.1, 1.2, 1.3, 1.35, 1.4] +
    [np.sqrt(2), np.sqrt(2) + 0.01] +
    [1.5, 1.6, 1.7, 1.8, 1.9, 2.0]
)
v0 = 0

x0_values = [(u0, v0) for u0 in u0_values]

for simulation in simulations:
    plot_poincare_sections(*simulation, x0_values)
```
